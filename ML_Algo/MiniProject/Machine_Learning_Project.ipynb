# **1. DECISION TREE**

```python
# Load libraries
import pandas as pd
# Import Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
# Import train_test_split function
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
import numpy as np 

dataset = pd.read_csv("/content/sample_data/winequality-red.csv")
dataset['quality'] = np.where(dataset['quality'] > 6, 1, 0)
dataset['quality'].value_counts()
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values

dataset.drop_duplicates(inplace=True)
print(dataset.isna().sum())
from sklearn.preprocessing import StandardScaler
std = StandardScaler()

#Standardize data set except dependent value(Quality feature)
dataset.iloc[:,:-1]=std.fit_transform(dataset.iloc[:,:-1])

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = 
train_test_split(X, Y, test_size=0.2, random_state=50)

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
Y_pred = clf.predict(X_test)
print("\nPredicted test data is : ")
print(Y_pred)

# Model Accuracy
print("\nAccuracy:",metrics.accuracy_score(Y_test, Y_pred)*100)

print("\nConfusion matrix:")
print(metrics.confusion_matrix(Y_test, Y_pred))
```

# **2. RANDOM FOREST**

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv("/content/sample_data/winequality-red.csv")
# input
X=dataset.iloc[:,:-1].values

# output
y=dataset.iloc[:,-1].values

dataset.drop_duplicates(inplace=True)
dataset.isna().sum()
dataset['quality'] = np.where(dataset['quality'] > 6, 1, 0)
dataset['quality'].value_counts()

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = 
train_test_split(X, y, test_size = 0.2, random_state = 50)

from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, random_state=0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

print("\nPrediction for test data set will be : ")
print(y_pred)

from sklearn.metrics 
import classification_report, confusion_matrix, accuracy_score
print('\nConfusion Matrix:')
print(confusion_matrix(y_test,y_pred))

print('\nClassification Report:')
print(classification_report(y_test,y_pred))

accuracy4=accuracy_score(y_test, y_pred)*100
print('\nAccuracy:',accuracy4)
```

# **3. LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
# Loading data
dataset = pd.read_csv('/content/sample_data/winequality-red.csv')  
# Create feature and target arrays
X = dataset.iloc[:, 0:7].values
y = dataset.iloc[:, 11].values 
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.18, random_state=50)

model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
model_score = (model.score(X_test, y_test)*100)

print("\nPrediction for test data set will be:")
print(y_predict)

print("\nAccuracy = ", model_score)

print("\nConfusion Matrix : ")
print(metrics.confusion_matrix(y_test, y_predict))
```

# **4. K-NEAREST NEIGHBORS**

```python
# Import necessary modules
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd  
# Loading data
dataset = pd.read_csv('/content/sample_data/winequality-red.csv')  

# Create feature and target arrays
X = dataset.iloc[:, 0:7].values
y = dataset.iloc[:, 11].values  

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state=50) 
knn = KNeighborsClassifier(n_neighbors=7)  
knn.fit(X_train, y_train)  

# Predict on dataset which model has not seen before
print("Predictions for test  data set are : ")
print(knn.predict(X_test))

# Calculate the accuracy of the model
print("\nAccuracy : ",knn.score(X_test, y_test)*100)

print('\nClassification Report : ')
print(classification_report(y_test,knn.predict(X_test)))
```

# **5. SUPPORT VECTOR MACHINE**
