```python
# importing required modules
import random
import csv

#initializing the attributes as a 2D array
attributes = [['Sunny','Rainy'],
              ['Warm','Cold'],
              ['Normal','High'],
              ['Strong','Weak'],
              ['Warm','Cool'],
              ['Same','Change']]

# finding no. of attributes type
num_attributes = len(attributes)


print (" \n Initial General Hypothesis : ['?','?','?','?','?','?']\n")
print ("\n Initial Specific Hypothesis : ['0','0','0','0','0','0']\n")

a = []
print("\n Given Data Set \n")

with open('/content/sample_data/EnjoySport.csv', 'r') as csvFile:
    reader = csv.reader(csvFile)
    for row in reader:
        a.append (row)
        print(row)


print("\n The initial value of hypothesis: ")
# initializing all values of array with 0
hypothesis = ['0'] * num_attributes
print(hypothesis)

# Comparing with First Training Example 
for j in range(0,num_attributes):
        hypothesis[j] = a[0][j];

# Comparing with Remaining Training Examples of Given Data Set

print("\n Find S: Finding a Maximally Specific Hypothesis :\n")

for i in range(0,len(a)):
  # if the training sample is positive
    if a[i][num_attributes]=='Yes':
            for j in range(0,num_attributes):
                if a[i][j]!=hypothesis[j]:
                  # initialize with most general hypothesis
                    hypothesis[j]='?'
                else :
                  # initialize with previous example value
                    hypothesis[j]= a[i][j] 
    print(" For Training Example No :{0} the hypothesis is : ".format(i),hypothesis)
                
print("\n The Maximally Specific Hypothesis for a given Training Examples :\n")
print(hypothesis)
```
